```python
url2= "https://www.ptt.cc/bbs/Beauty/M.1464224357.A.269.html"
res = requests.get(url2)
soup = bs(res.text) 

for i in soup.select('.push'):
    print '推噓   :',i.select('.hl.push-tag')[0].text
    print '回文者 :',i.select('.f3.hl.push-userid')[0].text
    print '回文   :',i.select('.f3.push-content')[0].text[1:].strip()
    print '時間   :',i.select('.push-ipdatetime')[0].text.strip()
    print '----------------------------------'
```

```python
h_count = 0 #噓
p_count = 0 #推
t_count = 0 #總
for i in soup.select('.push'):
    t_count = t_count + 1
    if i.select('.hl.push-tag')[0].text.strip().encode('utf-8') == '噓':
        h_count = h_count + 1
    elif i.select('.hl.push-tag')[0].text.strip().encode('utf-8') == '推':
        p_count = p_count + 1
print "鄉民評價:",
if p_count > h_count:
    print "好評"
else:
    print "負評"
print "-----整理-----"
print '推文 :', p_count
print '噓文 :', h_count
print '箭頭 :', t_count-(h_count+p_count)
print '總回應數 :', t_count
```

```python
import requests
from lxml.etree import tostring ,HTML
```

```python
url = 'http://rate.bot.com.tw/xrt/quote/l6m/JPY'
rs = requests.get(url)
ele = HTML(rs.text)
```

```python
import datetime
x=[]
y1=[]
y2=[]
for i in ele.xpath('//tr')[2:]:
    time = datetime.datetime.strptime(str(i.xpath('td[@class="text-center"]/a/text()')[0]), "%Y/%m/%d")
    x.append(time)
    y1.append(i.xpath('td[@class="rate-content-cash text-right print_table-cell"]/text()')[0])
    y2.append(i.xpath('td[@class="rate-content-cash text-right print_table-cell"]/text()')[1])
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.plot( x,y1, label='buys')
plt.plot(x, y2, label='sells')
plt.legend()
plt.show()
```

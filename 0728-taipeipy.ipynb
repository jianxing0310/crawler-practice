```python
import numpy as np 
import matplotlib.pyplot as plt 
X = np.linspace(-np.pi, np.pi, 256) 
cosx = np.cos(X) 
sinx = np.sin(X) 
plt.xlim(-4.0, 4.0) 
plt.ylim(-1.0, 1.0) 
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi]) 
plt.yticks([-1, 0, 1]) 
plt.plot(X, cosx, 'b-', linewidth=1.0) 
plt.plot(X, sinx, 'r-', linewidth=1.0)
```

```python
n = 1024 
X = np.random.normal(0, 1, n) 
Y = np.random.normal(0, 1, n) 
T = np.arctan2(Y,X) 
plt.axes([0.025,0.025,0.95,0.95]) 
plt.scatter(X,Y, s=75, c=T, alpha=.5) #alpha 透明度
plt.xlim(-1.5,1.5) 
plt.xticks([]) 
plt.ylim(-1.5,1.5) 
plt.yticks([]) 
```

```python
# 長條圖 迴圈 找出最小值 標示出來
```

```python
n = 7 
X = np.arange(n) 
Y1 = [0.5, 0.6, 0.65, 0.7, 0.8, 0.55, 0.9] 
plt.bar(X, Y1, facecolor='#9999ff') 
for x,y in zip(X,Y1): 
    if y == sorted(Y1)[1]: 
        plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom') #標示
        plt.ylim(0, 1.25)
```

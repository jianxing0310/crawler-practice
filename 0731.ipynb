# 互動


```python
from ipywidgets import interact
```

```python
def f(x):
    print(x)
```

```python
f('hi')
```

```python
interact(f, x=2);
```

```python
interact(f, x=(1,5));
```

```python
interact(f, x=3.);
```

```python
interact(f, x="you");
```

```python
interact(f, x={'a':'apple', 'o':'orange'});
```
